```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_nan_ratio(clean_df):
    ratios = []
    for col_ in clean_df.columns:
        #print(col_, ': ')
        total_rows = clean_df[col_].shape[0]
        ratio = 100*(total_rows-clean_df[col_].isna().sum())/total_rows
        ratios.append(ratio)

    _ = plt.hist(ratios, bins=20)  # arguments are passed to np.histogram
    plt.title("Histogram with 'auto' bins")
```

# INOUT

## drop non-descrip

```python
df = pd.read_csv('/content/INOUT.csv')
df.columns.shape
```

```python
inout_drop_cols = [
    'NAME',
    'HELP',
    'REQUESTDATE',
    'ADDRESS_ID',
    'ABVIOUSTYPE',
    'PC_ACTIVITIES_ID',
    'M_PIINOUT_REQUEST_ID',
    'RESPONSIBLEBPARTNER',
    'RESPONSIBLEMANAGER',
    'FROMUSERMANAGERID',
    'COM_BPARTNER_ID_F',
    'DOCUMENTNO1',
    'ACCT_AC_JOURNAL_ID',
]
```

```python
clean_df = df.drop(inout_drop_cols, axis=1)
clean_df.head()
```

```python
clean_df.to_csv('/content/meaningfull/INOUT.csv', index=False)
```

## keep descriptive features:

```python
desc_cols = [
    'CREATED',
    'UPDATED',
    'C_DOCSTATUS_ID',
    'DOCUMENTDATE',
    'ACCEPTANCETYPE'
]
```

```python
plot_nan_ratio(clean_df=clean_df[desc_cols])
```
